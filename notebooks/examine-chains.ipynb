```python
import numpy as np
import os
import pickle
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
test_dir = '../tests/test-mcmc-run/'
run_name = 'marg-disk-n12-sig3'

pkl_file = os.path.join(test_dir, run_name, 'chain-probabilities.pkl')

with open(pkl_file, 'rb') as f:
    data = pickle.load(f)
```

```python
data['prior'].shape
```

```python
burn_in = 100
indx = 0
prior = data['prior'][burn_in:,indx]
like = data['likelihood'][burn_in:,indx]

plt.plot(prior, label='prior')
plt.plot(like, label='likelihood')
plt.xlabel('Sample')
plt.ylabel('Log probability')
plt.legend()
```
